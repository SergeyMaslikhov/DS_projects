```python
from sentence_transformers import SentenceTransformer, InputExample, losses, evaluation
from torch.utils.data import DataLoader
from sklearn.model_selection import train_test_split
import numpy as np
import pandas as pd
```

```python
df = pd.read_excel('train_sample1.xlsx')
X_train, X_test, y_train, y_test = train_test_split(df[['target', 'anchor']], df['score'], stratify=df['score'], random_state=42)
#Define your train examples.
X_train_anch = X_train['anchor'].values
X_targ = X_train['target'].values
train_examples = []
test_examples = []
for i in range(len(X_train)):
  # print([df.loc[i, 'target'], df.loc[i, 'score']])
  train_examples.append(InputExample(texts=[X_train_anch[i], X_targ[i]], label=float(y_train.values[i])))
evaluator = evaluation.EmbeddingSimilarityEvaluator(X_test['target'].values, X_test['anchor'].values, np.float16(y_test.values))
#Define your train dataset, the dataloader and the train loss
train_dataloader = DataLoader(train_examples, shuffle=True, batch_size=16)

# evaluator = evaluation.EmbeddingSimilarityEvaluator(sentences1, sentences2, scores)
```

```python
#Define the model. Either from scratch of by loading a pre-trained model
pre_trained = ['sentence-transformers/multi-qa-MiniLM-L6-cos-v1',
               'cointegrated/rubert-tiny',
               'bert-base-multilingual-cased'
               ]
model = SentenceTransformer('cointegrated/rubert-tiny')
train_loss = losses.CosineSimilarityLoss(model)
model.fit(train_objectives=[(train_dataloader, train_loss)], epochs=60, warmup_steps=100, save_best_model=True,evaluator=evaluator, output_path='rubert-tiny')
```
